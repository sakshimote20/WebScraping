### defining the dynamic name

```python
ref=100000
for i in range(1,11):
    print('A'+str(i+ref)[1:])
```

### defining the image downloader function

```python
def download(link,img_path):
    res=requests.get(link)   # sending the request
    fd=open(img_path,'wb')   # getting image in wb format
    fd.write(res.content)    # saving the image
    fd.close()
```

### downloading all the images

```python
paths=[]
img_ids=[]
ref=100000
i=1
for link in tqdm( df['img_link']):
    
  img_id = "A" + str(i+ref)[1:]  # creating img id
  i+=1
  img_path='imgs/'+ img_id + '.jpg' # creating image path
  download(link,img_path)          # downloading the image
  paths.append(img_path) 
  img_ids.append(img_id)
```

### saving ids and paths in DataFrame

```python
df['img_id']=img_ids
df['img_path']=paths
df.head()
```

### saving the dataset

```python
df.to_csv('imgdataset.csv',index=False)
```
